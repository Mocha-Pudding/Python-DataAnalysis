## 【项目01】  商铺数据加载及存储

### 作业要求：
- 1、成功读取“商铺数据.csv”文件
- 2、解析数据，存成列表字典格式：[{'var1':value1,'var2':value2,'var3':values,...},...,{}]
- 3、数据清洗：
 - ① comment，price两个字段清洗成数字
 - ② 清除字段缺失的数据
 - ③ commentlist拆分成三个字段，并且清洗成数字
- 4、结果存为.pkl文件

```python
# 读取数据   G:\Python学习\Python-DataAnalysis\CLASSDATA_Chapter00基础语言\商铺数据.csv

import os
work_path = os.getcwd()
file_path = work_path + "\商铺数据.csv"
# print(file_path)
f = open(file_path,'r',encoding='utf-8')
for line in f.readlines()[:10]:
    print(line.split(','))
```

```python
f.seek(0)
# 评论 comment
for line in f.readlines()[:10]:
    print(line.split(',')[2])

f.seek(0)
# 价格 price
for line in f.readlines()[:10]:
    print(line.split(',')[4])

f.seek(0)
# 评论列表 commentlist
for line in f.readlines()[:10]:
    print(line.split(',')[-1])

f.seek(0)
# 评论列表 commentlist
for line in f.readlines()[:10]:
    print(line.split(',')[-1].split('                                '))
```

```python
# 清洗数据

# ① comment，price两个字段清洗成数字
# 清洗评论
def comment_func(s):
    if '条' in s:
        return (int(s.split(' ')[0]))
    else:
        return ('缺失数据')

# 清洗价格
def price_func(s):
    if '￥' in s:
        return(int(s.split('￥')[-1]))
    else:
        return('缺失数据')

# ③ commentlist拆分成三个字段，并且清洗成数字
# 拆分commentlist
def commentlist_func(s):
    if len(s) == 3:
        quality = float(s[0][2:])
        environment = float(s[1][2:])
        service = float(s[2][2:])
        return ([quality,environment,service])
    else:
        return ('缺失数据')
    
f.seek(0)
for line in f.readlines()[:10]:
    comment = comment_func(line.split(',')[2])
    print(comment)

print('-'*80)
f.seek(0)
for line in f.readlines()[:10]:
    price = price_func(line.split(',')[4])
    print(price)
    
print('-'*80)
f.seek(0)
for line in f.readlines()[:10]:
    commentlist = commentlist_func(line.split(',')[-1].split('                                '))
    print(commentlist)
```

```python
# 整体数据清洗
# 'classify', 'name', 'comment', 'star', 'price', 'address', 'commentlist'

datalist = []
n = 0

f.seek(0)

for i in f.readlines()[1:]:
    data = i.split(',')
    classify = data[0]
    name = data[1]
    comment = comment_func(data[2])
    star = data[3]
    price = price_func(data[4])
    address = data[5]
    qua = commentlist_func(data[-1].split('                                '))[0]
    env = commentlist_func(data[-1].split('                                '))[1]
    ser = commentlist_func(data[-1].split('                                '))[2]
#     print(classify,name,comment,star,price,address,qua,env,ser)
    
    if "缺失数据" not in [comment, price, qua]:
        n += 1
        data_list = [['classify', classify],
                    ['name', name],
                    ['comment', comment],
                    ['star', star],
                    ['price', price],
                    ['address', address],
                    ['quality', qua],
                    ['environment', env],
                    ['service', ser]]
#         print(data_list)
        datalist.append(dict(data_list))
        print('成功读取%i条数据' %n)
        print(dict(data_list))
    else:
        continue
        
f.close()
print('-'*120)
print(datalist)
print("数据格式化完成!")
print("总共加载%i条数据！" %n)
```

```python
# 复合列表 构建 字典
lst1 = ['a','b','c','d','e','f']
lst2 = [1,2,3,4,5,6]
datalist_test = []
m = 0
for i in range(len(lst1)):
    m += 1
    alpha = lst1[i]
    num = lst2[i]
    datalist_test1 = [['alpha',alpha],
                     ['number',num]]
#     print(datalist_test1)
    print("读取第%d条数据：" %m)
    print(dict(datalist_test1))
    datalist_test.append(dict(datalist_test1))
print('-'*60)
print(datalist_test)
```

```python
# 结果存为.pkl文件

import pickle

path = work_path + "\data.pkl"
pickle_file = open(path,'wb')
pickle.dump(datalist, pickle_file)
pickle_file.close()
print('完成！')
```
